```python

#P1
#running pandas dan numpy
import pandas as pd  
import numpy as np
#menampilkan array 1-dimensi
ds = pd.Series([2, 4, 6, 8, 10])
print(ds)
```

```python
#P2
data = pd.Series(['a','b','c']) #!sS
np1 = np.array(['x','y','z'])
data_np = pd.Series(np1)
print(data)
print(data_np)
#mengatur indeks pada numpy array data-np di atas
data_np = pd.Series(np1, index=[2,3,4])
print(data_np)
```

```python
#P3 contoh operasi
ds1 = pd.Series([2, 4, 6, 8, 10])
ds2 = pd.Series([1, 3, 5, 7, 9])
dsa = ds1 + ds2  #operasi jumlah
dsb = ds1 - ds2
dsc = ds1 * ds2
dsd = ds1 / ds2
print(dsa)
print(dsb)
print(dsc)
print(dsd)
```

```python
#P4 konver pandas module Series ke python list
import pandas as pd
ds = pd.Series([2, 4, 6, 8, 10])
print("Pandas Series dan type")
print(ds)
print(type(ds))
print("Konver Pandas Series ke Python list")
print(ds.tolist())
print(type(ds.tolist()))
```

```python
#P5 convert a NumPy array to a Pandas series
import numpy as np
import pandas as pd
np_array = np.array([0, 20, 30, 40, 50])
print("NumPy array:")
print(np_array)
new_np = pd.Series(np_array)
print("konversi Pandas series:")
print(new_np)
```

```python
# P6 https://numpy.org/doc/1.16/reference/generated/ numpy.random.RandomState.html

import pandas as pd
import numpy as np
#generate angka random
num_state = np.random.RandomState(100)  
num_series = pd.Series(num_state.normal(10, 4, 20))
print(num_series)
#print(num_state)
```

```python
#P15. Metode apply untuk menjalankan fungsi pada setiap baris pada data series
#Metode apply menjalankan fungsi pada baris data series
import pandas as pd
data = [1,1,2,3,5,8,13,21]
d_int = pd.Series(data)
def kecilbesar(num):
    if num <3:
      return "kecil"
    elif num >=3 and num <13:
      return "sedang"
    else:
      return "besar"
    
d_int.apply(kecilbesar)
```
